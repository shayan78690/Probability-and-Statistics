```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# Two types of probability distribution
# 1. Probability Mass Function
# 2. Probability Density Function
```

```python
# Probability Density Function
```

```python
# Normal distribution >> It had some mu and sigma
# majority of datapoint is clustered around mean
mu = 0
sigma = 0.1

s = np.random.normal(mu, sigma, size = 10000)
```

```python
len(s) # it means you want 1000 samples
```

```python
plt.hist(s ,30)
plt.show()
```

```python
sns.distplot(s) # or
```

```python
# Uniform distribution >> all the outcomes has equal probability of occur in different range
u = np.random.uniform(0, 1, 10000)
plt.hist(u)
plt.show()
```

```python
sns.distplot(u) # or
```

```python
# Log normal distribution
l = np.random.lognormal(mean = 0, sigma = 0.5, size = 1000)
plt.hist(l, bins = 100)
plt.show()
```

```python
# Log normal distribution
l = np.random.lognormal(mean = 0, sigma = 0.5, size = 1000)
l = l[np.where(abs(l) < 5)]
plt.hist(l, bins = 100)
plt.show()
```

```python
sns.boxplot(l) # outlier checking
# since log normal is right skewd so outliers are present
```

```python
sns.distplot(l)
```

```python
# lognormal becomes normal distribution if you take log of the values
sns.distplot(np.log(l))
```

```python
# square root transformation
sns.distplot(np.sqrt(l))
```

```python
# Chi-Square Distribution
c = np.random.chisquare(df = 30, size = 10000)
c
```

```python
c = np.random.chisquare(df = 30, size = 10000)
c = c[np.where(abs(c) < 100)] # for visible range
```

```python
plt.hist(c, bins = 1000)
plt.show()
```

```python
# multiple chi-square distribution in one plot
c = np.random.chisquare(df = 30, size = 10000)
d = np.random.chisquare(df = 10, size = 10000)
c = c[np.where(abs(c) < 100)]
d = d[np.where(abs(d) < 100)]
plt.hist(c, bins = 1000)
plt.hist(d, bins = 1000)
plt.show()
```

```python
# T-Distribution
t = np.random.standard_t(df = 3, size = 10000)
t = t[np.where(abs(t) < 100)]
plt.hist(t, bins = 1000)
plt.show()
```

```python
# if you increase dof
# T-Distribution
t = np.random.standard_t(df = 10, size = 10000)
t = t[np.where(abs(t) < 100)]
plt.hist(t, bins = 1000)
plt.show()
```

```python
# F-Distriibution
f = np.random.f(dfnum = 100, dfden = 100, size = 10000)
f = f[np.where(abs(f) < 10)]
plt.hist(f, bins = 1000)
plt.show()
```

```python
# Bernoulli distribution >> binomial distribution with one trial is bernoulii distribution
b = np.random.binomial(n = 1, p = 0.5, size  = 1000)
plt.hist(b)
plt.show()
```

```python
# Binomial Distribution
b = np.random.binomial(n = 100, p = 0.5, size  = 1000)
plt.hist(b)
plt.show()
```

```python
# Poisson's Distribution
p = np.random.poisson(lam = 5, size = 1000)
plt.hist(p, bins = 20)
plt.show()
```
