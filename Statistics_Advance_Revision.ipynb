```python
import numpy as np
import matplotlib.pyplot as plt

# Creating a series of data in range of 1-50
X = np.linspace(1, 50, 100)

# Creating a function for normal distribution
def normal_dist(x, mean, sd):
    prob_density = (1 / (sd * np.sqrt(2 * np.pi))) * np.exp(-0.5*((x - mean) / sd) ** 2)
    return prob_density

# Calculating mean and standard deviation
mean = np.mean(X)
sd = np.std(X)

# Apply the function to the data
pdf = normal_dist(X, mean, sd)

# Plotting the data
plt.figure(figsize = (6, 4))
plt.title("Normal Distribution")
plt.plot(X, pdf, color = 'red')
plt.xlabel("Values of random variable X")
plt.ylabel("Probability")
plt.show()
```

#Write a Python function to calculate the probability density function of a normal distribution with given mean and standard deviation at a given point.


```python
import math

def normal_pdf(x, mean, sd):
  variance = sd **2
  exponential = -((x - mean) **2) / (2 * variance)
  cofficient = 1 / math.sqrt(2 * math.pi * variance)
  pdf = cofficient * math.exp(exponential)
  return pdf
normal_pdf(5,2,1)
```

#Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4 and plot a histogram of the results using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a random sample of size 1000 from a Binomial distribution
n = 1000  # Sample size
p = 0.4  # Probability of success
sample = np.random.binomial(n, p, size=1000)

# Plot a histogram of the results using Fredman diaconis rule i.e. bins='fd'
plt.figure(figsize=(6,3))
plt.hist(sample,bins='fd',density=True, edgecolor='black')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution (n=1000, p=0.4)')
plt.show()
```

#Write a Python function to calculate the cumulative distribution function of a Poisson distribution with given mean at a given point

```python
import math

def poisson_cdf(mean, point):
    cdf = 0
    for k in range(point + 1):
        cdf = cdf + math.exp(-mean) * pow(mean, k) / math.factorial(k)
    return cdf


#Calculating cdf for mean =6 ,x=2
result = poisson_cdf(mean=6,point=2)
print(f'Poison cdf for mean = 6 at x = 2 is {result}')


#Verifying the above value by importing poission from scipy.stats
from scipy.stats import poisson
print(f'Poison cdf using scipy.stats library for mean = 6 at x = 2 is {poisson.cdf(2,mu=6)}')
```

#Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the sample mean and variance.

[ ]


```python
import numpy as np

# Set the random seed for reproducibility
np.random.seed(4)

#Lam: Known number of occurrences
#Size: The shape of the returned array
samp = np.random.poisson(lam=5,size=1000)

sample_mean = np.mean(samp)
sample_variance = np.var(samp)

print("Poission Sample Mean :", sample_mean)
print("Poission Sample Variance :", sample_variance)
```
